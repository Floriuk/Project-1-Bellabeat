```python
df_users = df1
```

```python
df_users = df_users[['Id','TotalSteps','ActivityDate']]
```

```python
# define the bins and labels for the activity levels
bins = [0, 3000, 6000, 9000, 12000, 30000]
labels = ['E', 'D', 'C', 'B', 'A']

# create a new column 'activity_level' with the labels based on the 'Total Steps' column
df_users['activity_level'] = pd.cut(df_users['TotalSteps'], bins=bins, labels=labels)

# print the updated DataFrame
print(df_users.head())
```

```python
df8_users = df8
df8_users['SleepDay'] = pd.to_datetime(df8['SleepDay']).dt.strftime('%m/%d/%Y')
df_users['ActivityDate'] = pd.to_datetime(df_users['ActivityDate']).dt.strftime('%m/%d/%Y')
df8_users = df8_users.rename(columns={'SleepDay': 'Date'})
df_users = df_users.rename(columns={'ActivityDate': 'Date'})
```

```python
relationship_users_steps_sleep = pd.merge(df_users,df8_users, on=(['Id','Date']), how = 'left')
```

```python
import seaborn as sns

# create scatter plot
sns.scatterplot(x='TotalMinutesAsleep', y='TotalSteps', data=relationship_users_steps_sleep, hue='activity_level', alpha=0.5)

# calculate correlation coefficient
corr_coef = relationship_users_steps_sleep['TotalMinutesAsleep'].corr(relationship_users_steps_sleep['TotalSteps'])

# add correlation coefficient to plot
plt.annotate('Correlation coefficient = {:.2f}'.format(corr_coef), 
             xy=(0.5, 0.9), xycoords='axes fraction', fontsize=12)

# add axis labels and title
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Total Steps')
plt.title('Relationship between Total Minutes Asleep and Total Steps')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
# display plot
plt.show()

```

```python
# get unique activity levels
activity_levels = relationship_users_steps_sleep['activity_level'].unique()
activity_levels = activity_levels.dropna()

# create a scatter plot for each activity level
for activity_level in activity_levels:
    # filter the data for the current activity level
    data = relationship_users_steps_sleep[relationship_users_steps_sleep['activity_level'] == activity_level]
    
    # create a scatter plot for the current activity level
    sns.scatterplot(x='TotalMinutesAsleep', y='TotalSteps', data=data, alpha=0.5)
    
    # calculate correlation coefficient for the current activity level
    corr_coef = data['TotalMinutesAsleep'].corr(data['TotalSteps'])
    
    # add correlation coefficient to plot for the current activity level
    plt.annotate('Correlation coefficient = {:.2f}'.format(corr_coef), 
                 xy=(0.5, 0.9), xycoords='axes fraction', fontsize=12)
    
    # add axis labels and title for the current activity level
    plt.xlabel('Total Minutes Asleep')
    plt.ylabel('Total Steps')
    plt.title('Relationship between Total Minutes Asleep and Total Steps for {}'.format(activity_level))
    
    # display plot for the current activity level
    plt.show()
```

```python
df5
```

```python
df_activity_time = pd.merge(df5,df6, on=(['Id','ActivityHour']), how = 'left')
```

```python
df_activity_time
```

```python
# convert datetime column to datetime dtype
df_activity_time['ActivityHour'] = pd.to_datetime(df_activity_time['ActivityHour'])

# add new column with hour extracted from datetime column
df_activity_time['Hour'] = df_activity_time['ActivityHour'].apply(lambda x: x.hour)
```

```python
df_activity_time
```

```python
df_activity_mean = df_activity_time.groupby('Hour')['Calories'].mean().reset_index()
df_activity_mean
```

```python
# create a box plot per hour using seaborn
# sns.boxplot(x='Hour', data=df_activity_time)
# sns.set_style("whitegrid")
# plt.xlabel('Hour')
# plt.ylabel('Calories')
# plt.title('Boxplot of Calories by Hour')
# plt.show()
```

```python
# Define the ranges and corresponding tags
ranges = [0, 80, 200, 10000]
tags = ['C', 'B', 'A']
# Add a new column 'Tag' based on 'Calories' ranges
df_activity_time['ActivityLevel'] = pd.cut(df_activity_time['Calories'], bins=ranges, labels=tags)
df_activity_time
```

```python
# Create a figure and subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 15))
# Flatten the axes array
axes = axes.flatten()
# Iterate over the activity levels and create a histogram for each subplot
activity_levels = df_activity_time['ActivityLevel'].unique()
for i, activity_level in enumerate(activity_levels):
    # Select data for the current activity level
    data = df_activity_time[df_activity_time['ActivityLevel'] == activity_level]
    # Create a histogram for the current subplot if i is within the bounds of the axes array
    if i < len(axes):
        sns.histplot(data=data, x='Hour', bins=10, ax=axes[i])
        axes[i].set_title(f'Activity Level: {activity_level}')
        axes[i].set_xlabel('Hour')
# Adjust the spacing between subplots
plt.tight_layout()
# Display the figure
plt.show()
```

```python
df_activity_time = df_activity_time.reset_index()
```

```python
df_compare= df_activity_time.groupby('Id')['Calories'].sum().reset_index()
df_compare['Rank'] = df_compare['Calories'].rank(ascending=False)
df_ranking = pd.merge(df_activity_time, df_compare, on='Id', how='left')
df_ranking = df_ranking.sort_values('Calories_y')
df_ranking
```

```python
df_ranking_c = df_ranking[df_ranking['ActivityLevel']=='C']
pivot_table = df_ranking_c.pivot_table(index='Hour', columns='Id', aggfunc='size', fill_value=0)

# Sort pivot table by sum of activities per user
pivot_table = pivot_table.loc[:, pivot_table.sum().sort_values(ascending=False).index]

sns.heatmap(pivot_table, cmap='Blues')
plt.xlabel('User')
plt.ylabel('Hour')
plt.title('Activity Heatmap C')
plt.show()
```

```python
df_ranking_b = df_ranking[df_ranking['ActivityLevel']=='B']
pivot_table = df_ranking_b.pivot_table(index='Hour', columns='Id', aggfunc='size', fill_value=0)

# Sort pivot table by sum of activities per user
pivot_table = pivot_table.loc[:, pivot_table.sum().sort_values(ascending=False).index]

sns.heatmap(pivot_table, cmap='Blues')
plt.xlabel('User')
plt.ylabel('Hour')
plt.title('Activity Heatmap B')
plt.show()
```

```python
df_ranking_a = df_ranking[df_ranking['ActivityLevel']=='A']
pivot_table = df_ranking_a.pivot_table(index='Hour', columns='Id', aggfunc='size', fill_value=0)

# Sort pivot table by sum of activities per user
pivot_table = pivot_table.loc[:, pivot_table.sum().sort_values(ascending=False).index]

sns.heatmap(pivot_table, cmap='Blues')
plt.xlabel('User')
plt.ylabel('Hour')
plt.title('Activity Heatmap A')
plt.show()
```
